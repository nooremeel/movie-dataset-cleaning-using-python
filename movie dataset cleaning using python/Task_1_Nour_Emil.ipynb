```python
#Read CSV file and store it in movie_data variable
movie_data=pd.read_csv('movie.csv')
movie_data
```

```python
#store the data into a data frame
movie_df = pd.DataFrame(movie_data)
movie_df.head()
```

```python
#remove empty cells
clean_movie_df=movie_df.dropna()
clean_movie_df
```

```python
#replace empty values
clean_movie_df = movie_df.fillna('Null')
clean_movie_df
```

```python
#replace values only in gross with mean value
gross_mean = movie_df['gross'].mean()
clean_movie_df['gross'] = movie_df['gross'].fillna(gross_mean)
clean_movie_df
```

```python
#replace values only in gross with median value
gross_mean = movie_df['gross'].median()
clean_movie_df['gross'] = movie_df['gross'].fillna(gross_mean)
clean_movie_df
```

```python
#replace values only in gross with mode value
#it doesnt work because data has no mode
gross_mean = movie_df['gross'].mode()
clean_movie_df['gross'] = movie_df['gross'].fillna(gross_mean)
clean_movie_df
```

```python
#convert num_critic_for_reviews to numeric same can be applied to other columns (note: the format is correct this is just an example)
movie_df = pd.to_numeric(movie_df['num_critic_for_reviews'])
```

```python
#fixing wrong data
movie_df.loc[0,'director_facebook_likes']=50
movie_df
```

```python
#replacing values
for x in movie_df.index:
    if movie_df.loc[x,'duration']>240:
        movie_df.loc[x,'duration']=240
```
